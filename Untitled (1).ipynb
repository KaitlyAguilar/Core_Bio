```python
cols = ['Sales Order Billed In Oracle', 'Customer Date Created', 'Sales Order Date Billed', 'Date Fulfilled', 'Time Fulfilled', 'Date Delivered', 'Time Delivered', 'Sales Order Date',
       'Walk In Flag', 'In Catalog Flag', 'Virtual Flag', 'Stocked Flag', 'Vendor Inactive Flag', 'Small Business Flag']
str_cols = IPPS_BI_Core_Dat.drop(cols, axis = 1).columns.tolist()
```

```python
def fix_mix_dtypes(lst, dic, typ):
    for col in lst:
        dic[col] = typ
```

```python
mix_col_date = ['Customer Date Created', 'Sales Order Date Billed', 'Date Fulfilled', 'Time Fulfilled', 'Date Delivered', 'Time Delivered', 'Sales Order Date']
mix_col_bool = ['Sales Order Billed In Oracle', 'Walk In Flag', 'In Catalog Flag', 'Virtual Flag', 'Stocked Flag', 'Vendor Inactive Flag', 'Small Business Flag']
dtype_dic = {}

fix_mix_dtypes(mix_col_date, dtype_dic, 'datetime64[ns]')
fix_mix_dtypes(mix_col_bool, dtype_dic, 'bool')
fix_mix_dtypes(str_cols, dtype_dic, 'str')

len(dtype_dic)
```

```python
IPPS_BI_Core_Dat[mix_col_bool].head()
```

```python
for col, dtype in dtype_dic.items():
    if 'datetime' in str(dtype):
        IPPS_BI_Core_Dat[col] = pd.to_datetime(IPPS_BI_Core_Dat[col], errors='coerce')
    elif dtype == 'bool':
        IPPS_BI_Core_Dat[col] = IPPS_BI_Core_Dat[col].astype('boolean')
    elif dtype == 'str':
        IPPS_BI_Core_Dat[col] = IPPS_BI_Core_Dat[col].astype('string')
```

```python
#drop Alternative Catalog Number, don't think it will create much of a difference and missing more than 50% of data
IPPS_BI_Core_Dat = IPPS_BI_Core_Dat.drop(columns = ['Alternative Catalog Number'])
```

```python
#(feel free to change the stuff below)
#Also got rid of names and emails
#Customer type is boolean?
#Got rid of a lot of duplicate columns, ie vendor, category, location. All have same/similar identification data.
#Have no idea if we should keep Product code, as it is similar to Product ID but also different? 
#drop irrelevant columns:
irr_columns = ['Category Prefix', 'Location ID Link', 'OFC Vendor Name', 'Vendor Name', 'Vendor Prefix', 
              'Vendor ID', 'Delivered By', 'Fulfilled by', 'PI Last Name', 'PI First Name', 'Customer Type', 'Standardized_Requester_Name',
              'Standardized_Requester_Email', 'Requester Email', 'Customer Name', 'Sales Order Taken by', 'Requester Name', 'Sales Order Number',
              'Product Code', 'PI Email']

IPPS_BI_Core_Dat = IPPS_BI_Core_Dat.drop(columns = irr_columns)
```

```python
IPPS_BI_Core_C = IPPS_BI_Core_Dat
IPPS_BI_Core_C.head()
```

```python
#categorical cols replace with unknown
string_cols = IPPS_BI_Core_C.select_dtypes(include='string').columns.tolist()

for col in string_cols:
    IPPS_BI_Core_C[col] = IPPS_BI_Core_C[col].astype('string').fillna('unknown')
```

```python
IPPS_BI_Core_C.isna().sum()
```

```python
IPPS_BI_Core_C['Line Item Number']
```

```python
for col in IPPS_BI_Core_C.columns:
    if IPPS_BI_Core_C[col].dtype == 'float64' or IPPS_BI_Core_C[col].dtype == 'int64':
        IPPS_BI_Core_C[col] = IPPS_BI_Core_C[col].fillna(IPPS_BI_Core_C[col].median())  # could also try mean
    elif IPPS_BI_Core_C[col].dtype == 'boolean':
        IPPS_BI_Core_C[col] = IPPS_BI_Core_C[col].fillna(False)
    elif pd.api.types.is_datetime64_any_dtype(IPPS_BI_Core_C[col]):
        IPPS_BI_Core_C[col] = IPPS_BI_Core_C[col].fillna(method='ffill')
```

```python
# Total_spent
customer_spending = IPPS_BI_Core_Churned.groupby('Customer ID')['Sales Order Total'].sum().reset_index()
customer_spending.columns = ['Customer ID', 'total_spent']
IPPS_BI_Core_Churned = IPPS_BI_Core_Churned.merge(customer_spending, on='Customer ID', how='left')
```
